```
 import nltk
 from nltk import word_tokenize, pos_tag, pos_tag_sents
 import pandas as pd
 df = pd.read_csv('Test(1334).csv', sep=',')
 df['Text']
 nltk.download('averaged_perceptron_tagger')
 nltk.download('punkt')
 texts = df['Text'].tolist()
 tagged_texts = pos_tag_sents(map(word_tokenize, texts))
 tagged_texts

 df['POS'] = tagged_texts
 df
```

```
texts = df['Text'].tolist()
tagged_texts = pos_tag_sents(map(word_tokenize, texts))
df['POS Tag'] =  tagged_texts

### Creation set
tagged_texts
tags = set()
for sentence in tagged_texts:
  for i in sentence:
    tags.add(i[1])
tags = list(tags) 


## Tag count
final_list = []
i=0
for sentence in df['POS Tag']:
  tag_count = {}
  for i in sentence:
    if i[1] not in tag_count:
      tag_count[i[1]] = 1 
    else:
      tag_count[i[1]] +=1
  final_list.append(tag_count)
df['Tag Count'] =  final_list

for n, i in enumerate(final_list):
  df = df.combine_first(pd.DataFrame(i, index=[n+0]))


#### Rearrange columns

cols = ['Text','POS Tag']
for i in tags:
  cols.append(i)

df  = df[cols]
print (df)
```
